| Column     | Description              |
|------------|--------------------------|

| `'AREA NAME'` | The 21 Geographic Areas or Patrol Divisions are also given a name designation that references a landmark or the surrounding community that it is responsible for. | `'Vict Descent'` | Victim's descent:<ul><li>`A` - Other Asian</li><li>`B` - Black</li><li>`C` - Chinese</li><li>`D` - Cambodian</li><li>`F` - Filipino</li><li>`G` - Guamanian</li><li>`H` - Hispanic/Latin/Mexican</li><li>`I` - American Indian/Alaskan Native</li><li>`J` - Japanese</li><li>`K` - Korean</li><li>`L` - Laotian</li><li>`O` - Other</li><li>`P` - Pacific Islander</li><li>`S` - Samoan</li><li>`U` - Hawaiian</li><li>`V` - Vietnamese</li><li>`W` - White</li><li>`X` - Unknown</li><li>`Z` - Asian Indian</li> |

| `'LOCATION'` | Street address of the crime. |

```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

```python
# Hour that has highest frequency crime
crimes["HOUR"] = crimes["TIME OCC"].str[:2].astype(int)
peak_crime_hour = crimes["HOUR"].value_counts().idxmax()
print(peak_crime_hour)
```

```python
# Area that has the largest frequency of night crimes (crimes committed between 10pm and 3:59am)?
crimes["TIME OCC"] = crimes["TIME OCC"].astype(int)
night_time = crimes[(crimes["TIME OCC"] > 2300) | (crimes["TIME OCC"] <= 359)]
peak_night_crime_location = night_time["AREA NAME"].value_counts().idxmax()  
print(peak_night_crime_location)
```

```python
# Identify the number of crimes committed against victims by age group (0-17, 18-25, 26-34, 35-44, 45-54, 55-64, 65+)
age_labels = ['0-17', '18-25', '26-34', '35-44', '45-54', '55-64', '65+']
age_bins = [0, 17, 25, 34, 44, 54, 64, float('inf')]
crimes['Age Group'] = pd.cut(crimes['Vict Age'], bins=age_bins, labels=age_labels)
victim_ages = crimes["Age Group"].value_counts().sort_index()
print(victim_ages)
```

```python
#creating a color palette for each age group
custom_palette = {'0-17': 'lightcyan', '18-25': 'powderblue', '26-34': 'skyblue','35-44':'lightblue', '45-54': 'steelblue', '55-64': 'lightsteelblue', '65+' : 'cornflowerblue'}
fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(x='Age Group', y='HOUR', data=crimes, palette=custom_palette)
#labels 
plt.xlabel("Age Group")
plt.ylabel("Hours of crime")
plt.title("Groups with ages 0-17 and 65+ have a smaller range of hours of crime", fontsize = 10)
plt.suptitle("Time Of Crime Per Victims' Age Group")

#show plot
plt.show()
```

```python
crimes_per_hour_per_age = crimes.groupby(["Age Group", "HOUR"]).size().reset_index(name='Number of Crimes')

# Assigning colors to each age group 
custom_palette = {'0-17': 'lightcyan', '18-25': 'powderblue', '26-34': 'skyblue', '35-44': 'lightblue', '45-54': 'steelblue', '55-64': 'lightsteelblue', '65+': 'cornflowerblue'}
sns.scatterplot(data=crimes_per_hour_per_age, x='HOUR', y='Number of Crimes', hue='Age Group', palette=custom_palette)

# Add labels and title
plt.xlabel('Crime Hour')
plt.ylabel('Number of Crimes')
plt.title('Number of Crimes per Hour by Age Group')

# Show the plot
plt.show()
```
